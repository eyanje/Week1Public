#### Explore the relationship between BMI and age. Does there appear to be an association?

```python
def plot_vars(x_var, y_var, title=None):
    fig, ax = plt.subplots()

    if title == None:
        title = f'{y_var} vs {x_var}'
    ax.set_title(title)

    ax.set_xlabel(x_var)
    ax.set_ylabel(y_var)

    ax.scatter(data[x_var], data[y_var])
    # Draw line of best fit?
    xs = np.arange(data[x_var].min(), data[x_var].max())

    ax.plot(xs, np.poly1d(np.polyfit(data[x_var], data[y_var], deg=1))(xs))
    fig.show()
```

```python
plot_vars('Age', 'BMI')
```

Fill in the function below. The function recieves a non-empty string and an int n, and returns a new string where the character at index n has been removed.Â¶


```python
def remove_char(old_string, n):
    return old_string[:n] + old_string[n + 1:]
```

Given a string, return a new string where the first and last characters have been exchanged.

```python
def swap_chars(old_string):
    return old_string[-1] + old_string[1:-1] + old_string[0]
```
